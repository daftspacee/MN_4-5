```python
"""Створіть одновимірний масив (вектор) зі значеннями n!, факторіала кожного натурального числа n, починаючи з 1, та виведіть його значення. """
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

n = 10  # Задайте бажану кількість елементів у векторі

factorial_vector = [factorial(i) for i in range(1, n + 1)]

# Виведення значень вектора факторіалів
print("Значення вектора факторіалів для перших", n, "натуральних чисел:")
print(factorial_vector)
```

```python
"""Створіть двовимірний масив (матрицю) розміром 3x3, 
заповніть його оберненими значеннями косинусів кожного натурального числа, починаючи з 1, та виведіть його значення. """

import numpy as np

# Функція для обчислення оберненого значення косинусу
def inverse_cosine(n):
    return 1 / np.cos(n)

# Створення двовимірного масиву розміром 3x3
matrix = np.empty((3, 3))

# Заповнення матриці оберненими значеннями косинусів
for i in range(3):
    for j in range(3):
        matrix[i, j] = inverse_cosine((i * 3) + j + 1)  # Обчислення оберненого косинусу для кожного елементу

# Виведення значень матриці
print("Значення матриці з оберненими значеннями косинусів:")
print(matrix)
```

```python
"""Створіть масив розміром 8x8, заповніть його випадковими дійсними числами 
від -1 до 1 та знайдіть відсоток чисел, які менше нуля та мають непарний індекс. """
import numpy as np

# Задання розміру матриці
size = (8, 8)

# Генерація масиву розміром 8x8 з випадковими дійсними числами від -1 до 1
random_matrix = np.random.uniform(-1, 1, size)

# Знаходження елементів, які менше нуля та мають непарний індекс
negative_odd_indices_count = np.sum((random_matrix < 0) & ((np.indices(size)[0] % 2 != 0) & (np.indices(size)[1] % 2 != 0)))

# Знаходження відсотку таких елементів відносно загальної кількості елементів
total_elements = size[0] * size[1]
percentage = (negative_odd_indices_count / total_elements) * 100

print(f"Відсоток чисел, менших за нуль та з непарним індексом: {percentage:.2f}%")
```

```python
"""Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю. """
import numpy as np

# Створення матриці розміром 3x3 з випадковими цілими числами
matrix = np.random.randint(1, 16, (3, 3))

print("Початкова матриця:")
print(matrix)

# Перевірка, чи можна знайти обернену матрицю
if np.linalg.det(matrix) != 0:  # Перевірка на невиродженість матриці
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
else:
    print("\nОбернена матриця не існує, оскільки вихідна матриця є виродженою.")
```
